The 2023 Arizona Diamondbacks were able to catch lightning in a bottle and produce one of the most unexpected playoff runs recently seen. Starting at the regular season, they were predicted to end either dead last in the most dominant division, finishing just above a 500 record, but more than 10 games behind the penciled-in favorite LA Dodgers. Then just barely passing to the playoffs, they were clear underdogs in every game played. Something hit right come playoffs, as they were able to go all the way into the World Series. Regardless of the outcome of the World Series, the 2023 as a whole was a success compared to the meager 2022 one.

A lot of things must have gone well to go from middle of the pack to World Series. This project will aim to highlight what specifically went right in 2023 where they went on to a record of 84-78, compared to a 2022 record of 74-88. Despite the loss of a huge offensive weapon like Daulton Varsho, the addition of Gabriel Moreno and Lourdes Gurriel Jr, helped the snakes' offense be enough to garner a ticket to the big dance. 


```python
import pandas as pd
import nfl_data_py as nfl
import matplotlib.pyplot as plt
from matplotlib import style

def ARI_Batting_Stats(season_year):
    url="https://www.baseball-reference.com/teams/ARI/"+season_year+".shtml/"
    reg_batting_df=pd.read_html(url, header=0 ,attrs={"id":"team_batting"})[0]
    reg_batting_df=reg_batting_df[pd.to_numeric(reg_batting_df['Rk'], errors='coerce').notnull()]
    reg_batting_df=reg_batting_df[~reg_batting_df["Pos"].isin(["P"])]
    reg_batting_df=reg_batting_df.set_index(["Name","Pos"])
    reg_batting_df=reg_batting_df.apply(pd.to_numeric)
    reg_batting_df=reg_batting_df[reg_batting_df.AB>=200]
    reg_batting_df=reg_batting_df.reset_index()
    reg_batting_df=reg_batting_df.set_index(["Name"])
    return reg_batting_df
```

```python
ARI_2022_Batting_df=ARI_Batting_Stats("2022").sort_values(["G","Pos"],ascending=False)
ARI_2023_Batting_df=ARI_Batting_Stats("2023").sort_values(["G","Pos"],ascending=False)
```

```python
ARI_2023_Batting_df.columns
```

```python
def batting_2023_metrics(metrics_to_analyze):
    metrics=metrics_to_analyze
    plt.rcParams["figure.figsize"] = [15, 10]
    x1=ARI_2023_Batting_df.index
    y1=ARI_2023_Batting_df[metrics[0]]
    y2=ARI_2023_Batting_df[metrics[1]]
    y3=ARI_2023_Batting_df[metrics[2]]
    plt.scatter(x1,y1,s=100, c='blue')
    plt.scatter(x1,y2,s=100, c='orange')
    plt.scatter(x1,y3,s=100, c='green')
    plt.title("2023 AZ Dbacks Top Batters Batting Metrics")
    plt.ylabel("Batting Metrics", fontsize=20)
    plt.legend(metrics)
    return
bat_neg2023=batting_2023_metrics(["SO","GDP","BB"])
```
